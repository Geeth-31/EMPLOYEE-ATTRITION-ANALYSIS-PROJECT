# Employee Attrition KPI Analysis

This notebook calculates key performance indicators (KPIs) from the cleaned IBM HR Employee Attrition dataset. These KPIs summarize workforce characteristics, employee attrition patterns, satisfaction metrics, and experience-related insights. The calculated KPIs will be used to support business insights and the Tableau dashboard.

## Importing Required Libraries

```python
# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")
```

## Loading the Cleaned Dataset

```python
df = pd.read_csv(r"C:\Users\Geeth\OneDrive\Documents\data analytics project\EMPLOYEE-ATTRITION-ANALYSIS-PROJECT\Data\Cleaned\IBM_HR_Employee_Attrition_Cleaned.csv")
print("Dataset loaded successfully.")
df.head()
```

# 1. Workforce KPIs

## KPI 1: Total Employees

```python
total_employees = len(df)
print("Total Employees:", total_employees)
```

## KPI 2: Employees Left

```python
employees_left = len(df[df["Attrition"] == "Yes"])
print("Employees Left:", employees_left)
```

## KPI 3: Employees Stayed

```python
employees_stayed = len(df[df["Attrition"] == "No"])
print("Employees Stayed:", employees_stayed)
```

## KPI 4: Overall Attrition Rate

```python
attrition_rate = (employees_left / total_employees) * 100
print("Overall Attrition Rate:",round( attrition_rate,2), "%")
```

## KPI 5: Average Employee Age

```python
average_age = df["Age"].mean()
print("Average Employee Age:", round(average_age, 2), "years")
```

## KPI 6: Average Monthly Income

```python
average_income = df["MonthlyIncome"].mean()
print("Average Monthly Income:", round(average_income, 2))
```

## KPI 7: Average Total Working Years

```python
average_working_years = df["TotalWorkingYears"].mean()
print("Average Total Working Years:", round(average_working_years, 2), "years")
```

# 2. Department KPIs

## KPI 8: Department-wise Attrition

```python
department_attrition = df[df["Attrition"] == "Yes"].groupby("Department").size()
print(department_attrition)
```

## Department-wise Attrition Rate

```python
department = df.groupby(["Department","Attrition"]).size().unstack()
department["Total Employees"] = department["Yes"] + department["No"]

department["Attrition Rate (%)"] = (department["Yes"] / department["Total Employees"]) * 100
department["Attrition Rate (%)"] = department["Attrition Rate (%)"].round(2)

department = department.sort_values(by="Attrition Rate (%)", ascending=False)
department
```

# 3. Job Role KPIs

## KPI 9: Job Role-wise Attrition

```python
jobrole_attrition = df[df["Attrition"] == "Yes"].groupby("JobRole").size()
print(jobrole_attrition)
```

## Job Role-wise Attrition Rate

```python
jobrole = df.groupby(["JobRole","Attrition"]).size().unstack()
jobrole["Total Employees"] = jobrole["Yes"] + jobrole["No"]

jobrole["Attrition Rate (%)"] = (jobrole["Yes"] / jobrole["Total Employees"]) * 100
jobrole["Attrition Rate (%)"] = jobrole["Attrition Rate (%)"].round(2)

jobrole = jobrole.sort_values(by="Attrition Rate (%)", ascending=False)
jobrole
```
