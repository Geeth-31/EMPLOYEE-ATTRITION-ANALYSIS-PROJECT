## Loading the Cleaned Dataset

```python
df = pd.read_csv(r"C:\Users\Geeth\OneDrive\Documents\data analytics project\EMPLOYEE-ATTRITION-ANALYSIS-PROJECT\Data\Cleaned\IBM_HR_Employee_Attrition_Cleaned.csv")
print("Dataset loaded successfully.")
df.head()
```

## KPI 1: Total Employees

```python
total_employees = len(df)
print("Total Employees:", total_employees)
```

## KPI 2: Employees Left

```python
employees_left = len(df[df["Attrition"] == "Yes"])
print("Employees Left:", employees_left)
```

## KPI 3: Employees Stayed

```python
employees_stayed = len(df[df["Attrition"] == "No"])
print("Employees Stayed:", employees_stayed)
```

## KPI 4: Overall Attrition Rate

```python
attrition_rate = (employees_left / total_employees) * 100
print("Overall Attrition Rate:", attrition_rate, "%")
```
