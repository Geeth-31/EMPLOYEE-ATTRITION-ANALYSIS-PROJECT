# 4. Correlation Analysis

```python
plt.figure(figsize=(14,10))
correlation = df.select_dtypes(include="number").corr()               #only numerical value columns
sns.heatmap(correlation, annot=True,fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# 5. Outlier Analysis

```python
numerical_columns = [ "Age", "DailyRate", "MonthlyIncome", "DistanceFromHome", "TotalWorkingYears"]
for col in numerical_columns:                                                        # meaningful numerical variables
    plt.figure(figsize=(6,4))                                      
    sns.boxplot(y=df[col])
    plt.title(f"Box Plot of {col}")
    plt.show()
```

# 6. Summary of Exploratory Data Analysis

### Key Findings

- The dataset contains 1,470 employee records and 31 features after data cleaning.
- No missing values or duplicate records were found in the dataset.
- Employees working overtime appeared to have comparatively higher attrition.
- Employees with lower monthly income and fewer total working years showed comparatively higher attrition.
- Employees living farther from the workplace appeared to have comparatively higher attrition.
- Research & Development was the largest department in the organization.
- Correlation analysis showed strong relationships among experience-related variables, while most other variables had weak to moderate correlations.
- Outliers were mainly observed in MonthlyIncome and TotalWorkingYears, indicating the presence of employees with exceptionally high salaries or extensive work experience.
