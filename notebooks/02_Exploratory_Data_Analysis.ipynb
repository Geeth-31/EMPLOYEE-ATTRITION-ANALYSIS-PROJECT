# Exploratory Data Analysis (EDA)

This notebook performs exploratory data analysis on the cleaned IBM HR Employee Attrition dataset to identify patterns, relationships, and factors associated with employee attrition.

#  Import Libraries

```python
# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")
```

# load dataset (cleaned)

```python
# Load the cleaned dataset
df = pd.read_csv(r"C:\Users\Geeth\OneDrive\Documents\data analytics project\EMPLOYEE-ATTRITION-ANALYSIS-PROJECT\Data\Cleaned\IBM_HR_Employee_Attrition_Cleaned.csv")
print("Dataset loaded successfully.")
df.head()                                          # to check first 5 rows in dataset
```

```python
print(df.shape)                                    # to check number of rows and columns in cleaned data
```

# 1. Target Variable Analysis

```python
# target variable is Attrition
df["Attrition"].value_counts()                          # to find number of Attrition values
```

```python
round(df["Attrition"].value_counts(normalize=True) * 100, 2)    # % of values in attrition
```

```python
# Visualization 
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="Attrition")
plt.title("Employee Attrition Distribution")

plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()
```
